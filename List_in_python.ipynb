#creating a list

```python
#empty
print([])
```

```python
#1D
print([1,2,3,4,5])#it is homogeneous
```

```python
#2D or nested list
print([1,2,3,[4,5]]) #it is heterogeneous
```

```python
#3D
print([[[1,2],[3,4]],[[5,6],[7,8]]])
```

```python
#heterogeneous
print(1,True,5.6,5+6j,'hello')
```

```python
#using type conversion
print(list('hello'))
```

#Accessing items from list

```python
#indexing
l=[1,2,3,4,5]
#positive index
print(l[0])
#negative index
print(l[-1])
```

```python
l=[1,2,3,[4,5]]
print(l[-1][-2])
```

```python
l=[1,2,3,[4,5]]
print(l[3][0])
```

```python
l=[[[1,2],[3,4]],[[5,6],[7,8]]]
print(l[1][0][1])
```

```python
print(l[0][0][1])
```

```python
#slicing
l=[1,2,3,4,5]
print(l[-3:])
```

```python
print(l[0::2])
```

```python
print(l[::-1])
```

#Adding item to a list

```python
#append
l=[1,2,3,4,5]
l.append(7)
print(l)
```

```python
l.append([7,8,9])
print(l)
```

```python
#extend
l=[1,2,3,4,5]
l.extend([7,8,9])
print(l)
```

```python
l=[1,2,3,4,5]
l.extend('hello')
print(l)
```

```python
#insert
l=[1,2,3,4,5]
l.insert(0,10)
print(l)
```

#editing items in list

```python
l=[1,2,3,4,5]
l[0]=10
print(l)
```

```python
l[1:4]=[50,40,30]
print(l)
```

#Deleting items from list

```python
#del
l=[1,2,3,4,5]
del l[0]
print(l)
```

```python
del l[1:3]
print(l)
```

```python
#remove
l=[1,2,3,4,5]
l.remove(3)
print(l)
```

```python
#pop
l=[1,2,3,4,5]
l.pop(0)
print(l)
```

```python
l.pop()
print(l)
```

```python
#clear
l=[1,2,3,4]
l.clear()
print(l)
```

#operations on lists

```python
#arithmetic(+,*)
l1=[1,2,3,4]
l2=[5,6,7,8]
print(l1+l2)
#concatenation/merge

```

```python
print(l1*3)
```

```python
#membership
l1=[1,2,3,4]
print(5 not in l1)
```

```python
print(5 in l1)
```

```python
#loop
l1=[1,2,3,4]
for i in l1:
  print(i)
```

#list functions

```python
#len/min/max/sorted
l1=[1,2,3,4]#homogenoues data type
print(len(l1))
print(min(l1))
print(max(l1))
print(sorted(l1))
print(sorted(l1,reverse=True))#it will create a new list it is not permanent
```

```python
#count
l=[1,2,1,3,4,1,5]
l.count(1)
```

```python
l.count(2)
```

```python
#index
l=[1,2,1,3,4,1,5]
l.index(1)#first occurence
```

```python
l.index(5)
```

```python
#reverse
l=[2,4,5,6]
#permanently reverse the list
l.reverse()
print(l)
```

```python
#sort(vs sorted)
l=[2,8,1,5]
print(l)
print(sorted(l))
print(l)
l.sort()
print(l)
```

```python
#copy -> shallow copy
l1=[1,2,3,4]
l2=l1.copy()
print(l1)
print(l2)
```
